For the exercises below, there are a few more turtle functions you might want to use.

* `penup` lifts the turtle's imaginary pen so it doesn't leave a trail when it moves.

* `pendown` puts the pen back down.

The following function uses `penup` and `pendown` to move the turtle without leaving a trail.

````python
from jupyturtle import penup, pendown

def jump(length):
    """Move forward length units without leaving a trail.

    Postcondition: Leaves the pen down.
    """
    penup()
    forward(length)
    pendown()
````

### Exercise

Write a function called `rhombus` that draws a rhombus with a given side length and a given interior angle. For example, here's a rhombus with side length `50` and an interior angle of `60` degrees.

````python
def rhombus( length: int, angle: int) :
    second_angle = 180 - angle
    for i in range(2):
        forward(length)
        left(angle)
        forward(length)
        left(second_angle)
````

You can use the following code to test your function.

````python
make_turtle()
rhombus(50, 60)
````

### Exercise

Now write a more general function called `parallelogram` that draws a quadrilateral with parallel sides. Then rewrite `rectangle` and `rhombus` to use `parallelogram`.

````python
def parallelogram( side1: int, side2: int, angle: float) -> None:
    second_angle = 180 - angle
    for _ in range(2):
        forward(side1)
        left(angle)
        forward(side2)
        left(second_angle)
````

````python
def rectangle( width: float, height: float):
    parallelogram( width, height, 90)
````

````python
def rhombus( length: float, angle: float):
    parallelogram( length, length, angle)
````

Hint: Write a function called `triangle` that draws one triangular segment, and then a function called `draw_pie` that uses `triangle`.

````python
def triangle( r: float, angle: float) :
    side = 2 * r * math.sin(math.radians(angle) / 2)
    base_angle = (180 - angle) / 2
    forward(r)
    left(180 - base_angle)
    forward(side)
    left(180 - base_angle)
    forward(r)


````

````python
def draw_pie( n: int, r: float) :
    angle = 360 / n
    for _ in range(n):
        triangle( r, angle)
        left(angle)
````

### Exercise

Write an appropriately general set of functions that can draw flowers like this.

![](https://github.com/AllenDowney/ThinkPython/raw/v3/jupyturtle_flower.png)

Hint: Use `arc` to write a function called `petal` that draws one flower petal.

````python
import math

def arc(r: float, angle: float):
    # Calculate the total length of the arc
    arc_length = 2 * math.pi * r * angle / 360
    
    # Calculate the number of tiny steps based on the arc length
    n = int(arc_length / 3) + 1 
    
    # Calculate how far to move and turn for each step
    step_length = arc_length / n
    step_angle = angle / n
    
    # Draw the arc by repeating the small steps
    for i in range(n):
        forward(step_length)
        left(step_angle)
````

````python
def petal( r: float, angle: float) :
    arc( r, angle)
    left(90)
    arc( r, angle)


def flower( n: int, r: float, angle: float) :
    for i in range(n):
        petal( r, angle)
        left(360 / n)
````

You can use the following code to test your functions.

````python
from jupyturtle import make_turtle, forward, left

def spiral(n, starting_length, angle, length_increase):
    """
    Draws a spiral.
    
    n: total number of steps/segments to draw
    starting_length: the length of the very first segment
    angle: degrees to turn left after every step
    length_increase: how much longer each step gets
    """
    length = starting_length
    
    for i in range(n):
        forward(length)
        left(angle)
        
        # Increase the length for the next step so the spiral grows outward
        length += length_increase

# Create the turtle and speed it up
make_turtle(delay=0)

# Draw a spiral with 150 steps, starting at length 1, 
# turning 15 degrees each time, and growing by 0.2 units per step.
spiral(n=150, starting_length=1, angle=15, length_increase=0.2)
````
